```python
import pandas as pd
```

### Import Data

```python
df = pd.read_csv('data/book_titles/good_reads_final.csv')
```

```python
df.head()
```

```python
books = df[['book_title', 'genre_1']]
```

```python
books.head()
```

### Get Genres

```python
books['genre_1'].value_counts()
```

```python
len(books['genre_1'].value_counts())
```

```python
pop_genres = books['genre_1'].value_counts().index.tolist()
```

```python
pop_genres # Genres in order of popularity
```

```python
pop_genres[:30]
```

```python
top_genres = pop_genres[:30]
```

```python
books
```

```python
romance = books.loc[books['genre_1'] == 'Romance']
```

```python
romance
```

```python
romance_list = list(romance['book_title'])
```

```python
with open('romance.txt', 'w') as f:
    [f.write(title.lstrip()) for title in romance_list]
```

### For each of the top 30 genres, write book titles to a txt file

```python
for genre in top_genres:
    genre_titles = books.loc[books['genre_1'] == genre]
    genre_list = list(genre_titles['book_title'])
    
    # Write to file
    filename = str(genre) + '.txt'
    with open(filename, 'w') as f:
        
        [f.write(title.lstrip()) for title in genre_list]
```
